# Murder

```python
M = N.dropna()
murder = M[M['OFNS_DESC'].dropna().str.contains("HOMICIDE")]
```

```python
murder['ARREST_BORO'].value_counts().plot.barh()
plt.title("Murder Arrests in boroughs")
plt.savefig('murder boro')
```

```python
murder.plot.scatter(x='Longitude', y='Latitude',c=murder['PERP_SEX'].map({'M':'r','F':'b'}),s=1)
```

```python
import folium
```

```python
map = folium.Map(location=(40.7128,-74.0060),zoom_start=10)

for each in murder.iterrows():
    marker = folium.CircleMarker(location=[each[1]['Latitude'],each[1]['Longitude']],
                          tooltip=each[1]['AGE_GROUP'],radius=3,color='red',opacity=.5,fill_color='red',)
    marker.add_to(map)

display(map)
```
